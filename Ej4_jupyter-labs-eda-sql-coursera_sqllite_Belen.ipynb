### Tarea 1

##### Mostrar los nombres de los sitios de lanzamiento únicos en la misión espacial

```python

# Consulta SQL para obtener los sitios de lanzamiento únicos
cur.execute('''SELECT DISTINCT "Launch_Site" FROM SPACEXTABLE;''')

# Recuperar los resultados
launch_sites = cur.fetchall()

# Mostrar los resultados
for site in launch_sites:
    print(site[0])



```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


##### Mostrar 5 registros donde los sitios de lanzamiento comienzan con la cadena 'CCA'

```python
# Consulta SQL para obtener 5 registros donde los sitios de lanzamiento comienzan con 'CCA'
cur.execute('''SELECT * FROM SPACEXTABLE WHERE "Launch_Site" LIKE 'CCA%' LIMIT 5;''')

# Recuperar los resultados
records = cur.fetchall()

# Mostrar los resultados
for record in records:
    print(record)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


##### Muestra la masa total de la carga útil transportada por los propulsores lanzados por la NASA (CRS)

```python
# Consulta SQL para sumar la masa total de las cargas útiles de las misiones "NASA (CRS)"
cur.execute('''
    SELECT SUM("PAYLOAD_MASS__KG_") 
    FROM SPACEXTABLE 
    WHERE "Customer" LIKE '%NASA (CRS)%';
''')

# Recuperar el resultado
total_mass = cur.fetchone()[0]

# Mostrar el resultado
print(f"Masa total de la carga útil transportada por los propulsores lanzados por la NASA (CRS): {total_mass} kg")
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


##### Muestra la masa media de la carga útil transportada por la versión de refuerzo F9 v1.1

```python
# Mostrar los valores únicos en la columna 'Landing_Outcome'
print(df['Landing_Outcome'].unique())
```

```python
# Consulta SQL para obtener la masa media de la carga útil de la versión de refuerzo "F9 v1.1"
cur.execute('''
    SELECT AVG("PAYLOAD_MASS__KG_") 
    FROM SPACEXTABLE 
    WHERE "Booster_Version" = 'F9 v1.1';
''')

# Recuperar el resultado
average_mass = cur.fetchone()[0]

# Mostrar el resultado
print(f"Masa media de la carga útil transportada por la versión de refuerzo F9 v1.1: {average_mass} kg")
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


##### Enumere la fecha en que se logró el primer resultado de aterrizaje exitoso en la plataforma de tierra.

_Hint:Usar function_ mínimo 

```python
# Consulta SQL para obtener la fecha del primer aterrizaje exitoso en la plataforma de tierra (RTLS)
cur.execute('''
    SELECT MIN(Date) 
    FROM SPACEXTABLE 
    WHERE "Landing_Outcome" = 'Success (ground pad)';
''')

# Recuperar el resultado
first_successful_rtls_date = cur.fetchone()[0]

# Mostrar el resultado
print(f"Fecha del primer aterrizaje exitoso en la plataforma de tierra: {first_successful_rtls_date}")
```

### Task 6
##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


##### Enumere los nombres de los propulsores que tienen éxito en la nave no tripulada y tienen una masa de carga útil superior a 4000 pero inferior a 6000

```python
cur.execute('''
    SELECT DISTINCT "Booster_Version"
    FROM SPACEXTABLE
    WHERE "Landing_Outcome" IN ('Success (drone ship)', 'Success')
    AND "PAYLOAD_MASS__KG_" > 4000
    AND "PAYLOAD_MASS__KG_" < 6000;
''')
# Recuperar el resultado
nombres_propulsores = cur.fetchall()

# Mostrar el resultado
print(f"Nombres de los propulsores que tienen éxito y tienen una masa de carga útil >4000 pero <6000: {nombres_propulsores}")
```

### Task 7




##### List the total number of successful and failure mission outcomes


##### Enumere el número total de resultados de misiones exitosas y fallidas

```python
# Obtener los valores únicos de la columna 'Mission_Outcome'
unique_mission_outcomes = df['Mission_Outcome'].unique()

# Mostrar los valores únicos
print(unique_mission_outcomes)
```

```python
# Ejecutar la consulta SQL para contar los resultados de misiones exitosas y fallidas
cur.execute('''
    SELECT "Mission_Outcome", COUNT(*) AS Total
    FROM SPACEXTABLE
    GROUP BY "Mission_Outcome";
''')

# Recuperar todos los resultados
resultados = cur.fetchall()

# Mostrar el resultado
print("Número total de resultados de misiones exitosas y fallidas:")
for resultado in resultados:
    print(f"{resultado[0]}: {resultado[1]}")

```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


##### Enumere los nombres de los booster_versions que han transportado la masa máxima de carga útil. Usar una subconsulta

```python
# Ejecutar la consulta SQL utilizando una subconsulta para encontrar los nombres de los booster_versions
cur.execute('''
    SELECT DISTINCT "Booster_Version"
    FROM SPACEXTABLE
    WHERE "PAYLOAD_MASS__KG_" = (
        SELECT MAX("PAYLOAD_MASS__KG_")
        FROM SPACEXTABLE
    );
''')

# Recuperar todos los resultados
nombres_boosters = cur.fetchall()

# Mostrar el resultado
print("Nombres de las versiones de booster que han transportado la masa máxima de carga útil:")
for booster in nombres_boosters:
    print(booster[0])
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


##### Enumere los registros que mostrarán los nombres de los meses, los landing_outcomes de fallas en la nave no tripulada, las versiones de refuerzo launch_site para los meses del año 2015.

**Nota: SQLLite no admite nombres de mes. Por lo tanto, debe usar substr(Date, 6,2) como mes para obtener los meses y substr(Date,0,5)='2015' para el año.**


```python
# Ejecutar la consulta SQL para obtener los registros requeridos
cur.execute('''
    SELECT 
        CASE 
            WHEN substr("Date", 6, 2) = '01' THEN 'January'
            WHEN substr("Date", 6, 2) = '02' THEN 'February'
            WHEN substr("Date", 6, 2) = '03' THEN 'March'
            WHEN substr("Date", 6, 2) = '04' THEN 'April'
            WHEN substr("Date", 6, 2) = '05' THEN 'May'
            WHEN substr("Date", 6, 2) = '06' THEN 'June'
            WHEN substr("Date", 6, 2) = '07' THEN 'July'
            WHEN substr("Date", 6, 2) = '08' THEN 'August'
            WHEN substr("Date", 6, 2) = '09' THEN 'September'
            WHEN substr("Date", 6, 2) = '10' THEN 'October'
            WHEN substr("Date", 6, 2) = '11' THEN 'November'
            WHEN substr("Date", 6, 2) = '12' THEN 'December'
        END AS "Month",
        "Landing_Outcome",
        "Booster_Version",
        "Launch_Site"
    FROM SPACEXTABLE
    WHERE "Landing_Outcome" LIKE 'Failure%'
      AND substr("Date", 0, 5) = '2015';
''')

# Recuperar todos los resultados
resultados = cur.fetchall()

# Mostrar el resultado
print("Registros de fallas en la nave no tripulada para el año 2015:")
for resultado in resultados:
    print(f"Month: {resultado[0]}, Landing Outcome: {resultado[1]}, Booster Version: {resultado[2]}, Launch Site: {resultado[3]}")
```

### Task 10




##### Clasifique el recuento de resultados de aterrizaje (como Fracaso (nave no tripulada) o Éxito (plataforma terrestre)) entre la fecha 2010-06-04 y 2017-03-20, en orden descendente.

```python
# Ejecutar la consulta SQL para contar los resultados de aterrizaje en el rango de fechas especificado
cur.execute('''
    SELECT 
        CASE 
            WHEN "Landing_Outcome" IN ('Success (ground pad)', 'Success (drone ship)', 'Success') THEN 'Éxito (plataforma terrestre)'
            WHEN "Landing_Outcome" LIKE 'Failure%' THEN 'Fracaso (nave no tripulada)'
            ELSE 'Otros' 
        END AS "Resultado",
        COUNT(*) AS "Total"
    FROM SPACEXTABLE
    WHERE "Date" BETWEEN '2010-06-04' AND '2017-03-20'
    GROUP BY "Resultado"
    ORDER BY "Total" DESC;
''')

# Recuperar todos los resultados
resultados = cur.fetchall()

# Mostrar el resultado
print("Recuento de resultados de aterrizaje entre 2010-06-04 y 2017-03-20:")
for resultado in resultados:
    print(f"Resultado: {resultado[0]}, Total: {resultado[1]}")
```
